```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
import seaborn as sns
style.use("ggplot")
%matplotlib.inline
```

```python
countries=gdp_data["Country"]    #create a variable of countries 
colors=('g','r','y')
```

```python
year_2000=gdp_data["2000"]
year_2001=gdp_data["2001"]
year_2002=gdp_data["2002"]
year_2003=gdp_data["2003"]
year_2004=gdp_data["2004"]
year_2005=gdp_data["2005"]
year_2006=gdp_data["2006"]
year_2007=gdp_data["2007"]
year_2008=gdp_data["2008"]
year_2009=gdp_data["2009"]
year_2010=gdp_data["2010"]
```

Pie chart- a pie visualization of countries gdp_per_capita

```python
plt.figure(figsize=(4,4),dpi=120,facecolor="#F5EEF8",edgecolor="#F9EBEA")
plt.pie(year_2000,labels=countries,colors=colors,
        shadow=True,
       wedgeprops={"edgecolor":"white"},
       startangle=90,
        explode=(0.04,0.04,0.02),
       autopct= "%1.0f%%",
        textprops={"fontsize":14},
       rotatelabels=True)
plt.title("GDP_per_capita of south america")
plt.tight_layout()
plt.show()
```

Horizontal barchart of countries vs gdp_per_capita

```python
plt.figure(figsize=(6,3),dpi=150,facecolor="#F5EEF6")          # a figure with dot per inch of 150,size 6x3 and a facecolor(with the colors's hexadecimal)
plt.barh(countries,year_2000,color="r",edgecolor="white",lw=0.5,label="countries")  # values to be plotted and plot properties
plt.tight_layout()
plt.title("Countries_GDP in 2000")
plt.xlabel("GDP per capita",rotation=42)
plt.xlim(0,2500)         #the limit of the x-axis that should be plotted
```

```python
fig,axes=plt.subplots(1,2,figsize=(16,6))
plt.tight_layout()
axes[0].bar(gdp_data["Country"],year_2001,label="2001",color="b",edgecolor="white",lw=3)
axes[0].set_xlabel("Countries")
axes[0].set_ylabel("gdp_per_capita")
axes[0].set_title("year 2000 Gdp_contribution",size=12)

axes[1].bar(gdp_data["Country"],year_2010,label="2010",color="r",edgecolor="white",lw=3)
axes[1].set_xlabel("Countries")

axes[1].set_title("year 2010 Gdp contribution",size=12)
```

```python
fig,axes=plt.subplots(1,2,figsize=(16,6),dpi=200)
plt.tight_layout()
sns.barplot(x="Country",y="2001",data=gdp_data,label="2001",ax=axes[0])   #a barplot using seaborn on axes 1 & 2
axes[0].set_xlabel("Countries")
axes[0].set_ylabel("gdp_per_capita")
axes[0].set_title("year 2000 Gdp_contribution",size=12)

sns.barplot(x="Country",y="2010",data=gdp_data,label="2010",ax=axes[1])
axes[1].set_xlabel("Countries")
axes[1].set_title("year 2010 Gdp contribution",size=12)
```
